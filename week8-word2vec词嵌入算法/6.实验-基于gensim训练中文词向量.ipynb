```python
import jieba
import codecs
import os


def read_file_and_cut(filepath: str, sents: list[list[str]]):
    """
    读取文件并分词
    """
    if os.path.exists(filepath) is False:
        return

    with codecs.open(filepath, 'r', 'utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            words = list(filter(lambda word: word.strip(), jieba.cut(line)))
            sents.append(words)


def load_files_to_sents(dirpath: str, ):
    """
    读取目录下的所有文件并分词
    """
    sents = []
    for file in os.listdir(dirpath):
        if file != '1.txt':
            continue
        filepath = os.path.join(dirpath, file)
        read_file_and_cut(filepath, sents)
    return sents


sents = load_files_to_sents('./data/')
```

```python
flattened_list = [item for sublist in sents for item in sublist]
uniq_words = set(flattened_list)
uniq_words
```

```python
from gensim.models import Word2Vec
from time import time

start_time = time()
model = Word2Vec(sents, min_count=2, vector_size=50)
word_vector = model.wv  # 词向量

# 遍历词向量
total = 0
word_vectors = set()
for word in word_vector.index_to_key:
    total += 1
    word_vectors.add(word)
    print(word, word_vector[word].shape)

uniq_words - word_vectors
```
