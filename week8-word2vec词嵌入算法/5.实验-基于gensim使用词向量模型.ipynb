```python
import gensim.downloader as api
from gensim.models.keyedvectors import Word2VecKeyedVectors


model_path = api.load("word2vec-google-news-300", return_path=True)

# 使用 word2vec-google-news-300 词向量模型
model = Word2VecKeyedVectors.load_word2vec_format(model_path, binary=True)
```

```python
# words = model.most_similar("husband", topn=5)  # 获取与 king 最相似的 5 个词

model.similarity("husband", "wife")  # husband 和 wife 的相似度
model.similarity('father', 'mother')  # father 和 mother 的相似度
```

```python
# 词向量可视化
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

words = ['man', 'woman', 'king', 'queen', 'cat', 'dog', 'mother', 'father']
vectors = np.array([model[word] for word in words])
pca = PCA(n_components=2)
vectors_2d = pca.fit_transform(vectors)

plt.figure()
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], edgecolors='k', c='r')
for word, (x, y) in zip(words, vectors_2d):
    plt.text(x, y, word)
plt.show()
```
