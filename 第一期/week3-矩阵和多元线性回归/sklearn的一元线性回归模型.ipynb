```python
import numpy
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import metrics # 计算平方误差
```

```python
x = [[50.0], [60.0],[70.0],[80.0],[90.0],[100.0],[110.0],[120.0]]
y = [280.0, 305.0, 350.0, 425.0, 480.0, 500.0, 560.0, 630.0]
```

```python
plt.scatter(x, y, color='red', marker='+')
```

```python
model = linear_model.LinearRegression()
model.fit(x, y)
model.intercept_, model.coef_
```

```python
import numpy as np

pre_y = model.predict(x)
metrics.mean_squared_error(y, pre_y)
```

```python
np.average((pre_y - y) ** 2)
```

```python
new_x = np.linspace(0, 150, 100)
new_y = model.predict(new_x.reshape(-1, 1))
plt.plot(new_x, new_y, color='blue')
plt.scatter(x, y, color='red', marker='+')
```
