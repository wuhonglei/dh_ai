```python
"""
使用 opencv 为图像添加高斯噪声 (即增加图像噪点)
"""

import os
import cv2
import numpy as np

img_dir = './misclassified_images/8'
img_files = os.listdir(img_dir)
for img_file in img_files:
    # 读取图片，添加高斯噪声，显示修改前后的图片
    img_path = os.path.join(img_dir, img_file)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    # 生成高斯噪声
    mean = 0
    std = 10
    gaussian_noise = np.random.normal(mean, std, img.shape).astype(np.float32)

    # 将噪声添加到图像
    noisy_img = img + gaussian_noise

    # 将像素值裁剪到0-255范围，并转换回8位图像
    noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)

    # 将原始图像和添加噪声后的图像拼接后显示
    img_concat = np.concatenate((img, noisy_img), axis=1)
    # 创建一个命名窗口
    cv2.namedWindow('img', cv2.WINDOW_NORMAL)
    # 设置窗口大小
    cv2.resizeWindow('img', 232, 116)
    cv2.imshow('img', img_concat)
    key = cv2.waitKey(0)
    if key == ord('q'):  # ESC
        cv2.destroyWindow('img')
        print('break')
        break
    print('continue')
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

import torch
import torchvision.transforms as transforms
import matplotlib.pyplot as plt


class AddGaussianNoise(object):
    def __init__(self, mean=0.0, std=1.0):
        self.mean = mean
        self.std = std

    def __call__(self, tensor):
        noise = torch.randn(tensor.size()) * self.std + self.mean
        return tensor + noise

    def __repr__(self):
        return self.__class__.__name__ + '(mean={0}, std={1})'.format(self.mean, self.std)


transform = transforms.Compose([
    transforms.ToTensor(),  # 将图像转换为张量
    AddGaussianNoise(mean=0.0, std=0.03),  # 添加高斯噪声
])


# 加载图像
def display_img(img_path):
    img = Image.open(img_path).convert('L')  # 假设使用灰度图像
    img_array = np.array(img)

    # 应用变换
    transformed_img = transform(img)

    # 转换为 NumPy 数组，便于显示
    transformed_img_np = transformed_img.numpy().squeeze()

    # 显示原图和带噪声的图像
    plt.figure(figsize=(4, 2))

    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(img, cmap='gray')

    plt.subplot(1, 2, 2)
    plt.title('Image with Gaussian Noise')
    plt.imshow(transformed_img_np, cmap='gray')

    plt.show()


img_dir = './misclassified_images/8'
img_files = os.listdir(img_dir)
for img_file in img_files:
    img_path = os.path.join(img_dir, img_file)
    display_img(img_path)
```
