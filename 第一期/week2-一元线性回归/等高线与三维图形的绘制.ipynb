```python
import matplotlib.pyplot as plt
import numpy as np

# 绘制三维坐标系
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
# ax2 = fig.add_subplot(122)

"""
绘制 3D 散点图 z = X**2 + Y**2
X 大小为 10 * 10 的矩阵
Y 大小为 10 * 10 的矩阵
Z 大小为 10 * 10 的矩阵
"""

x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

# ax1.scatter(x, y, z, c='r', marker='o')
# ax2.scatter(x, y, c='r', marker='o')
# ax1.plot_surface(x, y, z)
plt.contour(x, y, z)
```
