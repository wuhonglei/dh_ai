```python
# 数f(x) = x^2 + 4x + 5

import matplotlib.pyplot as plt
import numpy as np


def f(x):
    return x**2 + 4*x + 5


def grad_f(x):
    return 2 * x + 4


def get_f_x(point):
    k = grad_f(point)
    y = f(point)
    b = y - k * point
    return lambda x: k * x + b


x = np.linspace(-5, 4, 200)
grad_x = np.linspace(-1, 3, 100)
point = 1

plt.plot(x, f(x))
plt.plot(grad_x, get_f_x(point)(grad_x), 'r')
```
