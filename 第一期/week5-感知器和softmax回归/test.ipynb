```python
import numpy as np
import matplotlib.pyplot as plt


def step_function(x):
    # 定义阶跃函数
    return np.where(x >= 0, 1, 0)


# 生成数据
x = np.linspace(-10, 10, 1000)
y = step_function(x)

# 绘制阶跃函数
plt.plot(x, y)
plt.title("Step Function")
plt.xlabel("x")
plt.ylabel("H(x)")
plt.show()
```
