```python
import numpy as np


def f(x):
    # 函数 f(x) = x^2 +3 * x + 2
    return x ** 2 + 3 * x + 2


def df(x):
    return 2 * x + 3


x_arr = np.linspace(-6.5, 3.5, 1000)
y_f = f(x_arr)
y_df = df(x_arr)
```

```python
import matplotlib.pyplot as plt


plt.plot(x_arr, y_f, label='f(x) = x^2 +3 * x + 2', linewidth=5)
plt.plot(x_arr, y_df, label="f'(x) = 2x +3", linewidth=5)
plt.legend()  # 对图例进行显示
plt.grid(True)  # 显示网格
```

```python
import torch
x = torch.linspace(-6.5, 3.5, 1000, requires_grad=True)
y_f = f(x)
y_f
```

```python
y_f.sum().backward()
```

```python
y_df = x.grad.detach().numpy()
y_f = y_f.detach().numpy()
new_x = x.detach().numpy()
```

```python
x.grad == df(x)
```

```python
plt.plot(new_x, y_f, label='f(x) = x^2 +3 * x + 2', linewidth=5)
plt.plot(new_x, y_df, label="f'(x) = 2x +3", linewidth=5)
plt.legend()
plt.grid(True)
```

```python
import torch

# 创建需要梯度的张量
tensor = torch.linspace(0, 10, steps=5, requires_grad=True)

# 执行一些计算
result = tensor * 2

# 计算梯度
result.sum().backward()

# 打印梯度
print(tensor.grad)
```
