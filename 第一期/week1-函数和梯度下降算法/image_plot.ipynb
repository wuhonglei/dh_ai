```python
import matplotlib.pyplot as plt

# 创建 x 和 y 数据列表
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# 绘制折线图
plt.plot(x, y, linewidth=5, marker='o')

# 设置图表标题并给坐标轴加上标签
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

plt.show()
```

```python
import mpl_toolkits.axisartist as axisartist 

def draw_clear_board(left, right, down, up):
  """
  绘制一个坐标系, X 轴范围是 [left, right], Y 轴范围是 [top, bottom]
  """
  board = plt.figure()
  axis = axisartist.Subplot(board, 111)
  board.add_axes(axis) # 添加坐标轴

  axis.set_aspect(1) # 设置坐标轴比例
  axis.axis[:].set_visible(False) # 隐藏原始坐标轴

  axis.axis["x"] = axis.new_floating_axis(0, 0)
  axis.axis["y"] = axis.new_floating_axis(1, 0)

  axis.axis["x"].set_axisline_style("->") # 设置箭头
  axis.axis["y"].set_axisline_style("->")

  axis.set_xlim(left, right)
  axis.set_ylim(down, up)

  return axis

draw_clear_board(-8, 8, -8, 8)
plt.show()
```

```python
import numpy as np

def draw_linear(k, b, left, right):
  """
  绘制一条直线 y = kx + b
  """
  x = np.linspace(left, right, 100)
  y = k * x + b
  plt.plot(x, y)
```

```python
left, right = -8, 8
draw_clear_board(left, right, left, right)

draw_linear(1, 0, -8, 8)
draw_linear(2, 1, -8, 8)
draw_linear(0.5, 0, -8, 8)
```

```python
def draw_quadratic(a, b, c, left, right):
  """
  绘制一个二次函数 y = ax^2 + bx + c
  """
  x = np.linspace(left, right, 100)
  y = a * x**2 + b * x + c
  plt.plot(x, y)

left, right = -8, 8
draw_clear_board(left, right, left, right)
draw_quadratic(0.2, 1, 0, -8, 8)
```
