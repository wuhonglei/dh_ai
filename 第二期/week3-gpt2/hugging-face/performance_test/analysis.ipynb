```python
from dataset import WritingPromptsDataset
import matplotlib.pyplot as plt

test_dataset = WritingPromptsDataset(
    prompt_path="./writingPrompts/test.wp_source",
    story_path="./writingPrompts/test.wp_target",
)

train_dataset = WritingPromptsDataset(
    prompt_path="./writingPrompts/train.wp_source",
    story_path="./writingPrompts/train.wp_target",
)

test_prompt_token_len = []
test_story_token_len = []
for item in test_dataset:
    test_prompt_token_len.append(len(item['prompt'].split()))
    test_story_token_len.append(len(item['story'].split()))

train_prompt_token_len = []
train_story_token_len = []
for item in train_dataset:
    train_prompt_token_len.append(len(item['prompt'].split()))
    train_story_token_len.append(len(item['story'].split()))


# 创建 2x2 的子图布局
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 第一行左侧：测试集 prompt 长度分布
axes[0, 0].hist(test_prompt_token_len, bins=100, edgecolor='black')
axes[0, 0].set_title('Test Prompt Length Distribution')
axes[0, 0].set_xlabel('Length')
axes[0, 0].set_ylabel('Frequency')

# 第一行右侧：训练集 prompt 长度分布
axes[0, 1].hist(train_prompt_token_len, bins=100, edgecolor='green')
axes[0, 1].set_title('Train Prompt Length Distribution')
axes[0, 1].set_xlabel('Length')
axes[0, 1].set_ylabel('Frequency')

# 第二行左侧：测试集 story 长度分布
axes[1, 0].hist(test_story_token_len, bins=100, edgecolor='black')
axes[1, 0].set_title('Test Story Length Distribution')
axes[1, 0].set_xlabel('Length')
axes[1, 0].set_ylabel('Frequency')

# 第二行右侧：训练集 story 长度分布
axes[1, 1].hist(train_story_token_len, bins=100, edgecolor='green')
axes[1, 1].set_title('Train Story Length Distribution')
axes[1, 1].set_xlabel('Length')
axes[1, 1].set_ylabel('Frequency')

# 调整子图之间的间距
plt.tight_layout()

# 显示图表
plt.show()
```
