```python
from dataset import WritingPromptsDataset
import matplotlib.pyplot as plt

test_dataset = WritingPromptsDataset(
    prompt_path="./writingPrompts/test.wp_source",
    story_path="./writingPrompts/test.wp_target",
)
prompt_token_len = []
story_token_len = []
for item in test_dataset:
    prompt_token_len.append(len(item['prompt'].split()))
    story_token_len.append(len(item['story'].split()))

# 分别绘制 prompt 和 story 的 token 长度分布
plt.hist(prompt_token_len, bins=100, edgecolor='black')
plt.title('Prompt Length Distribution')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()

plt.hist(story_token_len, bins=100, edgecolor='black')
plt.title('Story Length Distribution')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    'prompt': prompt_token_len,
    'story': story_token_len
})

# 分析 prompt 的 token 长度分布
# df['prompt'].describe()
# 分析 story 的 token 长度分布
# df['story'].describe()
# 分析 prompt 和 story 的 token 长度分布
print(df.describe())

# 获取获取 prompt 的 90% 分位数
prompt_90_percentile = df['prompt'].quantile(0.9)
print(f"Prompt 的 90% 分位数: {prompt_90_percentile}")

# 获取获取 story 的 90% 分位数
story_90_percentile = df['story'].quantile(0.9)
print(f"Story 的 90% 分位数: {story_90_percentile}")
```
